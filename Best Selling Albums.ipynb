```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_best-selling_albums'
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
soup.find_all('table')
table = soup.find_all('table')[3]
table
```

```python
album_titles = table.find_all('th')
album_titles
```

```python
album_table_titles = [title.text.strip() for title in album_titles]
album_table_titles
```

```python
import pandas as pd
df = pd.DataFrame(columns = album_table_titles)

df
```

```python
column_row = table.find_all('tr')
for row in column_row[1:]:
    row_cells = row.find_all('td')
    data_cells = [data.text.strip() for data in row_cells]
    length = len(df)
    df.loc[length] = data_cells
```

```python
df
```

```python
df.to_csv(r'Desktop\albumsales.csv', index = False)
```
