```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.basketball-reference.com/leaders/pts_career.html'
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.text, 'html')
soup
```

```python
soup.find_all('table', class_="suppress_glossary suppress_csv sortable stats_table")
soup.find_all('table')[2]
table = soup.find_all('table')[2]
```

```python
table_title = table.find_all('th')
nba_table_title = [title.text.strip() for title in table_title]
nba_table_title
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = nba_table_title)
df
```

```python
data = table.find_all('tr')
for rows in data[1:]:
    row_data = rows.find_all('td')
    scoring_list = [data.text.strip() for data in row_data]
    length = len(df)
    df.loc[length] = scoring_list
```

```python
df
```

```python
df = df.apply(lambda x: x.str.strip())
```

```python
df.to_csv(r'Desktop\NBA TOP scorrers of all time.csv', index=False)
```
